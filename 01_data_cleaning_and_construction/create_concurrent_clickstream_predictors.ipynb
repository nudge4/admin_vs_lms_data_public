This script creates all of the early-term LMS predictors for concurrent courses in the target term, for all non-first-term observations.

```python
import pandas as pd
import numpy as np
df = pd.read_stata("~\\Box Sync\\Clickstream\\data\\full\\LMS_data_standardized_with_credit.dta")
```

```python
df.shape
```

```python
vccsid_sets = []
l = list(np.unique(df.vccsid))
n = len(l) // 2000 + 1
for i in range(2000):
    vccsid_sets.append(l[(n*i):(n*(i+1))])
```

```python
for i,ll in enumerate(vccsid_sets):
    print(i)
    ll_series = pd.DataFrame({'vccsid':ll})
    part = ll_series.merge(df, how='inner', on=['vccsid'])
    print(part.shape[0])
    part.to_stata("~\\Box Sync\\Clickstream\\data\\full\\concurrent\\part{}.dta".format(i),
                  write_index=False)
    print("")
```

```python
def compute_concurrent(t):
    new_t = t.groupby(['vccsid','strm','college','section','course']).agg('sum').reset_index()
    # for p in ["avg_click_cnt_per_session_qrt1", "avg_session_len_qrt1", "avg_time_per_day_qrt1", "cum_act_day_cnt_qrt1", "irreg_session_gap_qrt1", "irreg_session_len_qrt1", "longest_inact_qrt1", "tot_click_cnt_qrt1", "tot_time_qrt1"]:
    for p in ["avg_session_len_qrt1", "irreg_session_len_qrt1", "tot_click_cnt_qrt1", "tot_time_qrt1"]:
        new_t.loc[:,p] = new_t.loc[:,p]/new_t.credit
    # for p in ["assign_sub_cnt_am_qtr1", "assign_sub_cnt_day_qtr1", "assign_sub_cnt_pm_qtr1", "on_time_assign_share_qtr1"]:
    for p in ["on_time_assign_share_qtr1", "assign_sub_cnt_qtr1"]:
        new_t.loc[:,p] = new_t.loc[:,p]/new_t.loc[:,"has_"+p]
        new_t.loc[:,'has_'+p] = new_t.loc[:,'has_'+p].apply(lambda x: int(x > 0))
    return new_t.fillna(0)
```

```python
part_new_list = []
for a in range(2000):
    print(a)
    part = pd.read_stata("~\\Box Sync\\Clickstream\\data\\full\\updated\\concurrent\\part{}.dta".format(a))
#     predictors = ["avg_click_cnt_per_session_qrt1", "avg_session_len_qrt1", "avg_time_per_day_qrt1", "cum_act_day_cnt_qrt1", "irreg_session_gap_qrt1", "irreg_session_len_qrt1", "longest_inact_qrt1", "tot_click_cnt_qrt1", "tot_time_qrt1", 
#                   "has_assign_sub_cnt_am_qtr1", "has_assign_sub_cnt_day_qtr1", "has_assign_sub_cnt_pm_qtr1", "has_on_time_assign_share_qtr1", 
#                   "assign_sub_cnt_am_qtr1", "assign_sub_cnt_day_qtr1", "assign_sub_cnt_pm_qtr1", "on_time_assign_share_qtr1"]
    predictors = ["avg_session_len_qrt1", "irreg_session_len_qrt1", 
                  "tot_click_cnt_qrt1", "tot_time_qrt1", 
                  "has_on_time_assign_share_qtr1", "on_time_assign_share_qtr1",
                  "has_assign_sub_cnt_qtr1", "assign_sub_cnt_qtr1"]
    for p in predictors:
        part.loc[:,p] = part.loc[:,p] * part.credit
    all_vccsid_strm = part.loc[:,['vccsid','strm']].drop_duplicates()
    n = all_vccsid_strm.shape[0]
    temp_list = []
    for i in range(n):
        part_sub = all_vccsid_strm.iloc[i:(i+1),:].merge(part, on=['vccsid','strm'])
        indx = []
        indx_2 = []
        m = part_sub.shape[0]
        for j in range(m):
            indx.extend([k for k in range(m) if k != j])
            indx_2.extend([j]*(m-1))
        part_sub1 = part_sub.iloc[indx,:].loc[:,['credit'] + predictors]
        part_sub1.index = range(m*(m-1))
        part_sub2 = part_sub.iloc[indx_2,:].loc[:,['vccsid','strm','college','section','course']]
        part_sub2.index = range(m*(m-1))
        temp = pd.concat([part_sub2, part_sub1], axis=1)
        temp_list.append(temp)
    part_new = pd.concat(temp_list)
    part_new = compute_concurrent(part_new)
    part_new_list.append(part_new.copy())
```

```python
merged = pd.concat(part_new_list).drop(['credit'], axis=1)
merged = merged.rename(columns = {p:(p+"c") for p in predictors})
```

```python
merged.to_stata("~\\Box Sync\\Clickstream\\data\\full\\LMS_data_standardized_concurrent.dta", write_index=False)
```
